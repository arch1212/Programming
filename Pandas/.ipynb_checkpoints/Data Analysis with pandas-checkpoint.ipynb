```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df.set_index('Date', inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f'{region}_price25ma']]
    else:
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])
```

### P3: Groupby

```python
import pandas as pd
df = pd.read_csv('Minimum Wage Data.csv', encoding = "latin")
df.to_csv("minwage.csv", encoding = "utf-8")
```

```python
df = pd.read_csv("minwage.csv")
df.head()
```

```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year").sort_index()[['Low.2018']].rename(columns = {'Low.2018': name})
        
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year").sort_index()[['Low.2018']].rename(columns = {'Low.2018':name}))
        
act_min_wage.head()
```
