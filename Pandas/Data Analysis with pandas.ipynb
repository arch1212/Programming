### P2: Graphing and Visualization

```python
import pandas as pd
df = pd.read_csv('avocado.csv')
df["Date"] = pd.to_datetime(df["Date"])
albany_df = df[df['region'] == 'Albany']
```

```python
albany_df.set_index("Date", inplace = True)
```

```python
albany_df['AveragePrice'].plot()
```

```python
# moving average to smoothen out this level of fluctuation
albany_df["AveragePrice"].rolling(25).mean().plot()
```

```python
# something is wrong
# check with the index
albany_df.index
```

```python
albany_df.sort_index(inplace = True)
albany_df["AveragePrice"].rolling(25).mean().plot()
```

```python
# saving the moving average to a column
albany_df['price_25_ma'] = albany_df['AveragePrice'].rolling(25).mean()
```

```python
albany_df.head(3)
```

```python
# first twenty five are going to be Nan
albany_df.head(26)
```

```python
# drop any rows that have missing value
albany_df.dropna().head(3)
```

```python
# remove warnings
# because lateron we could modify df and pandas wants us to know that it might impact albany df so instead we create a
# a copy of df
albany_df = df.copy()[df['region'] == 'Albany']
albany_df.set_index('Date', inplace = True)
albany_df.sort_index(inplace = True)
albany_df['price_25_ma'] = albany_df['AveragePrice'].rolling(25).mean()
# albany_df.head(3)
albany_df.dropna().head(3) # not in place
```

```python
# easier way to do this
df['region'].unique()
```

```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region'] = region]
    region_df.set_index('Date', inplace = True)
    region_df.sort_index(inplace = True)
    region_df['price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[['price25ma']]
    else:
        graph_df = graph_df.join(region_df['price25ma'])
```
